```python
import pandas as pd
movies_file = "/content/movies.dat"
movies_df = pd.read_csv(movies_file, sep="::", engine='python', header=None, names=['MovieID', 'Title', 'Genres'])
ratings_file = "/content/ratings.dat"
ratings_df = pd.read_csv(ratings_file, sep="::", engine='python', header=None, names=['UserID', 'MovieID', 'Rating', 'Timestamp'])
users_file = "/content/users.dat"
users_df = pd.read_csv(users_file, sep="::", engine='python', header=None, names=['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code'])
print("Movies DataFrame:")
print(movies_df.head())
print("\nRatings DataFrame:")
print(ratings_df.head())
print("\nUsers DataFrame:")
print(users_df.head())
```

```python
merged_df = pd.merge(ratings_df, movies_df, on='MovieID', how='left')
final_df = pd.merge(merged_df, users_df, on='UserID', how='left')
print("Final DataFrame:")
print(final_df.head())
```
